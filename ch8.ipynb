### 8. 맷플롯립(matplotlib)
데이터를 시각화하는 모듈로, 그래프와 차트를 제공

```python
# 필요한 모듈 가져오기
import numpy as np
# 맷플롯립은 여러 모듈로 구성됨
# 그 중 그래프를 제공하는 모듈 꺼내기
import matplotlib.pyplot as plt
```

```python
# 그래프를 그릴 때 한글이 깨지는 현상이 발생
from matplotlib import rc
rc('font', family='Malgun Gothic')
```

#### 8.1 그래프 그리기
선 그래프
- plot: 선 그래프를 그리는 함수

```python
# 먼저 그래프에 담을 데이터 생성
a = np.arange(1,11)
a
# 숫자 배열을 그래프에 그리기
# plot(데이터)
plt.plot(a)
# x축: 우리가 담은 데이터
# y축: plt가 자동으로 설정함
```

#### 막대 그래프
- bar: 화면에 막대 그래프를 그리는 함수

```python
# 그래프에 담을 데이터 생성
# 학급별 평균 시험 점수
a = ['1반', '2반', '3반', '4반']
b = [80,90,85,95]

# x축: 학급
# y축: 점수
plt.bar(a, b)
```

산점도 
- scatter: 화면에 산점도를 그리는 함수

```python
# 산점도에 담을 데이터 생성
x = [30,25,10,50,25]
y = [9,3,10,6,2]

# 산점도 그리기
plt.scatter(x, y)
plt.title('산점도')
```

히스토그램
- hist: 화면에 히스토그램을 그리는 함수

```python
# 히스토그램에 사용 할 데이터 만들기
# 시험 점수 100개 만들기
scores = np.random.randint(50,101,100)
scores

# 히스토그램 그리기
# plt.hist(데이터, 구간)
plt.hist(scores, bins=5)

plt.title('시험 점수 분포')
plt.xlabel('시험 점수')
plt.ylabel('학생의 수')
```

#### 8.5 시본으로 여러 그래프 그리기
서브 플롯: 도화지 하나에 여러 그래프를 그리는 기능

- figure(피겨): 도화지(캔버스)
- axes(엑시스): 그래프가 그려지는 영역 (서브플롯이라고도 부름)

```python
df = pd.DataFrame( {'이름':['영희','철수','민수','지연'],
                    '점수':[85,90,80,100],
                    '반':['1반','1반','2반','2반']} )
```

```python
# 도화지 만들기
# subplots(행열의 크기, 캔버스의 크기)
fig, axs = plt.subplots(1, 2, figsize=(10,3))
# 엑시스는 행열에 따라 크기가 결정됨
sns.barplot(data=df, x='이름', y='점수', ax=axs[0])
axs[0].set_title('학생 점수(막대)')

# 두번째 영역에 그래프 그리기
sns.lineplot(data=df, x='이름', y='점수', ax=axs[1])
axs[1].set_title('학생 점수(선)')

# 글자가 겹치지 않게 레이아웃을 자동으로 조정
plt.tight_layout()
```

요일별 평균 기온데이터를 그래프로 그리기

```python
# 요일별 평균 기온 데이터
df = pd.DataFrame({'요일':['월','화','수','목','금','토','일'],
                   '기온':[12,14,13,15,17,18,16]})
# 피겨와 엑시스 만들기
# 엑시스는 1행 2열
fig, axs = plt.subplots(1, 2, figsize=(10,3))
# 첫번째 영역에 그래프 그리기
# 마지막에 그래프를 그릴 위치를 설정
sns.lineplot(data=df, x='요일', y='기온', ax=axs[0])
# 각 서브플롯에 타이틀 설정
axs[0].set_title('요일별 기온(선)')
# 두번째 영역에 그래프 그리기
sns.scatterplot(data=df, x='요일', y='기온', ax=axs[1])
axs[1].set_title('요일별 기온(산점도)')
```

```python
df = pd.DataFrame({'과일':['사과','바나나','포도','딸기'],
                   '판매량':[120,80,150,100]})
fig, axs = plt.subplots(1, 3, figsize=(14,4))
sns.barplot(data=df, x='과일',y='판매량', ax=axs[0])
axs[0].set_title('과일별 판매량(막대)')
sns.lineplot(data=df, x='과일',y='판매량', ax=axs[1])
axs[1].set_title('과일별 판매량(선)')
sns.scatterplot(data=df, x='과일',y='판매량', ax=axs[2])
axs[2].set_title('과일별 판매량(산점도)')
```

```python
df = pd.DataFrame( {'공부시간':[2,3,4,5,6,7,8],
                    '점수':[55,60,65,70,75,85,90],
                     } )
fig, axs = plt.subplots(3,1, figsize=(3,10))
sns.scatterplot(data=df, x='공부시간',y='점수',ax=axs[0])
axs[0].set_title('공부시간-점수(산점도)')
sns.barplot(data=df, x='공부시간',y='점수',ax=axs[1])
axs[1].set_title('공부시간-점수(막대)')
sns.lineplot(data=df, x='공부시간',y='점수',ax=axs[2], marker='o')
axs[2].set_title('공부시간-점수(그래프)')

plt.tight_layout()
```

```python
df1 = pd.DataFrame({'과일':['사과','바나나','포도','딸기'],
                   '판매량':[120,80,150,100]})
df2 = pd.DataFrame( {'이름':['영희','철수','민수','지연'],
                    '점수':[85,90,80,100],
                    } )
fig, axs = plt.subplots(2, 2, figsize=(8,8))
sns.barplot(data=df1, x='과일',y='판매량', ax=axs[0,0])
axs[0,0].set_title('과일별 판매량(막대)')
sns.lineplot(data=df1, x='과일',y='판매량', ax=axs[0,1], marker='o')
axs[0,1].set_title('과일별 판매량(선)')
sns.barplot(data=df2, x='이름',y='점수', ax=axs[1,0])
axs[1,0].set_title('학생별 점수(막대)')
sns.scatterplot(data=df2, x='이름',y='점수', ax=axs[1,1])
axs[1,1].set_title('학생별 점수(산점도)')


plt.tight_layout()
```
