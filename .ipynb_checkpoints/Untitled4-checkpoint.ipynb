#### 6.2 없는 데이터 nan 처리하기
- nan: csv에서 빈셀을 의미. 값이 없다

빈값이 있는지 확인하기
- isnull: 빈값이 있는지 확인
- isnull.sum: 빈값의 개수

```python
data = pd.read_csv('0917/서울특별시 마포구_법정동별_월별_코로나19 확진자 현황.csv', encoding='euc-kr')
data
df = data[['기준연도','기준월','법정동','확진자수']]
# 확진자수에 빈값이 있는지 확인
# 각 행에 대해서 검사 -> bool 시리즈 반환
df['확진자수'].isnull()
# 빈값 개수 세기
# bool -> int
# false: 0
# true: 1
df['확진자수'].isnull().sum()
```

빈값 삭제하기
- dropna: 빈값이 있는 행을 삭제
- dropna(subset=['컬럼']): 빈값이 있는지 특정 컬럼만 검사하여, 삭제

```python
# 빈값이 있는 행을 모두 삭제
df.dropna()
# 확진자 수가 없는 행만 삭제
df.dropna(subset=['확진자수'])
# dropna는 결과만 보여주고, 원본은 그대로
df
```

빈값을 다른값으로 대체하기
- fillna(새로운 값) : 컬럼 하나만 처리할 때
- fillna({ 컬럼1: 새로운 값, 컬럼2: 새로운 값 }) : 여러 컬럼을 한번에 처리 할 때

```python
# 변경된 내용을 원본에 넣기
# 법정동에 빈값이 있으면 '모름'으로 표시
df['법정동'].fillna('모름')
# 확진자수에 빈값이 있으면 0으로 표시
df['확진자수'] = df['확진자수'].fillna(0)
# 원본은 그대로
# 여러 컬럼을 한번에 교체
# 법정동과 확진자수의 빈값을 한번에 처리
df.fillna({'법정동':'모름', '확진자수':0})
df
# -> 원본은 그대로
```

#### 6.4 집계 및 변형
- groupby : 특정 컬럼을 기준으로 행 묶기
- 데이터를 그룹화 한 후에 합계, 평균을 구할 수 있다

```python
# 코로나 데이터 가져오기
data = pd.read_csv('0917/서울특별시 마포구_법정동별_월별_코로나19 확진자 현황.csv', encoding='euc-kr')
data
df = data[['기준연도','기준월','법정동','확진자수']]
df
```

```python
# 동별 총 확진자수 구하기
result = df.groupby('법정동').sum()
result['확진자수']
result['확진자수']['공덕동']
```
