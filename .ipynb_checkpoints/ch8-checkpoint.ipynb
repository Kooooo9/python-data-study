### 8. 맷플롯립(matplotlib)
데이터를 시각화하는 모듈로, 그래프와 차트를 제공

```python
# 필요한 모듈 가져오기
import numpy as np
# 맷플롯립은 여러 모듈로 구성됨
# 그 중 그래프를 제공하는 모듈 꺼내기
import matplotlib.pyplot as plt
```

```python
# 그래프를 그릴 때 한글이 깨지는 현상이 발생
from matplotlib import rc
rc('font', family='Malgun Gothic')
```

#### 8.1 그래프 그리기
선 그래프
- plot: 선 그래프를 그리는 함수

```python
# 먼저 그래프에 담을 데이터 생성
a = np.arange(1,11)
a
# 숫자 배열을 그래프에 그리기
# plot(데이터)
plt.plot(a)
# x축: 우리가 담은 데이터
# y축: plt가 자동으로 설정함
```

#### 막대 그래프
- bar: 화면에 막대 그래프를 그리는 함수

```python
# 그래프에 담을 데이터 생성
# 학급별 평균 시험 점수
a = ['1반', '2반', '3반', '4반']
b = [80,90,85,95]

# x축: 학급
# y축: 점수
plt.bar(a, b)
```

산점도 
- scatter: 화면에 산점도를 그리는 함수

```python
# 산점도에 담을 데이터 생성
x = [30,25,10,50,25]
y = [9,3,10,6,2]

# 산점도 그리기
plt.scatter(x, y)
plt.title('산점도')
```

히스토그램
- hist: 화면에 히스토그램을 그리는 함수

```python
# 히스토그램에 사용 할 데이터 만들기
# 시험 점수 100개 만들기
scores = np.random.randint(50,101,100)
scores

# 히스토그램 그리기
# plt.hist(데이터, 구간)
plt.hist(scores, bins=5)

plt.title('시험 점수 분포')
plt.xlabel('시험 점수')
plt.ylabel('학생의 수')
```

```python
a = np.arange(10, 21)
plt.plot(a)
```

```python
a = ['사과', '바나나', '포도', '오렌지']
b = [50,80,40,70]

plt.bar(a,b)
```

```python
x = [1,2,3,4,5]
y = [40,55,65,80,85]
plt.scatter(x,y)
plt.title('산점도 그래프')
plt.xlabel('공부 시간')
plt.ylabel('시험 점수')
```

```python
a = np.random.randint(0, 11, 100)
a
plt.hist(a)
plt.title('유튜브 시청 시간 분포')
plt.xlabel('시청 시간(시간)')
plt.ylabel('사람 수')
```

#### 8.2 여러 그래프 그리기
figure(피겨): 흰 도화지
- 도화지의 사이즈를 설정하거나, 여러 그래프를 한번에 그릴 때 사용


axes(엑시스): 도화지 안에 들어가는 그래프

```python
# 1부터 10까지 숫자 생성
a = np.arange(1,11)
# 선그래프 그리기
plt.plot(a)
# plot 함수를 쓰면 자동으로 figure(도화지)가 생성된다
```

```python
# 흰 도화지 만들기
# 가로 3인치 세로 3인치 만큼 크기 설정하기
plt.figure(figsize=(3,3))
# 도화지 위에 그래프 그리기
plt.plot(np.arange(1,11))
```

```python
# 도화지를 만들고 그 위에 그래프 여러 개 그리기
# 도화지 만들기
fig = plt.figure(figsize=(10,3))
# 도화지 위에 그래프를 어떤 모양으로 그릴지 생각하기
# 두개의 그래프를 한줄에 표시
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
students = [1,2,3,4,5] # 학번
scores = [80,90,70,85,95] # 점수
ax1.plot(scores)
ax2.bar(students, scores)
```

```python
# 학생 시험 점수 데이터

students = [1,2,3,4,5] # 학번
scores = [80,90,70,85,95] # 점수
```

```python
# 첫번째로 선 그래프 그리기
# 위에서 만든 엑시스 사용
ax1.plot(scores)
ax2.bar(students, scores)
```

```python
fig, axs = plt.subplots(1,2, figsize=(10,3))


students = [1,2,3,4,5] # 학번
scores = [80,90,70,85,95] # 점수

axs[0].plot(scores)
axs[1].bar(students, scores)
```

```python
fig = plt.figure(figsize=(10,3))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
fluits = ['사과','바나나','포도','오렌지','복숭아']
quantity = [50,80,40,70,60]
ax1.plot(quantity)
ax2.bar(fluits, quantity)
plt.title('과일 판매량')
```

```python
fig, axs = plt.subplots(1,2, figsize=(10,3))
day = ['월','화','수','목','금']
time = [2,3,4,2,5]
axs[0].plot(day, time)
axs[1].bar(day, time)
plt.title('요일별 유튜브 시청 시간')
```

```python
fig = plt.figure(figsize=(10,3))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

day = ['월','화','수','목','금']
time = [2,3,4,2,5]
ax1.plot(day, time)
ax2.bar(day, time)
plt.title('요일별 유튜브 시청 시간')
```

```python
fig = plt.figure(figsize=(10,3))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

day = ['월','화','수','목','금']
time = [1,0.5,2,1.5,2.5]
ax1.bar(day, time)
ax2.scatter(day, time)
```

```python
fig = plt.figure(figsize=(10,3))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

city = ['서울','부산','대구','광주','제주']
temp = [15.5,17.8,16.2,15.0,18.3]
ax1.plot(city, temp)
ax2.scatter(city, temp)
```
