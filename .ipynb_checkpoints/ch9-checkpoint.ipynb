#### 인천 초중교 학교 지도 그리기
공공데이터 포털에서 전국 학교 데이터를 가지고 온 다음에  
필요한 데이터만 선택해서 학교 위치를 지도 위에 그린다  

folium:지도 기능을 제공하는 모듈

```python
# 모듈 불러오기
import pandas as pd
import folium 
```

```python
data = pd.read_csv('sample/전국초중등학교위치표준데이터.csv', encoding='euc-kr')
```

```python
# 여러 컬럼을 선택
df = data[['학교명','시도교육청명','학교급구분','위도','경도']]
school = df[df['시도교육청명'] == '인천광역시교육청']
```

```python
map = folium.Map(location=[37.458077,126.715780], zoom_start=10)
```

```python
folium.Marker(location=[37.458077, 126.715780], tooltip='인천광역시').add_to(map)
```

```python
for i in school.index:
    name = school.loc[i, '학교명']
    typ = school.loc[i, '학교급구분']
    lat = school.loc[i, '위도']
    lon = school.loc[i, '경도']

    # 조건문으로 마커의 색 결정하기
    # 학교급구분에 따라 색 결정
    if typ=='초등학교':
        color = 'orange'
    elif typ=='중학교':
        color = 'green'
    elif typ=='고등학교':
        color = 'blue'

    # tooltip: 마우스를 올렸을 때만 정보가 표시됨
    # popup: 클릭하면 정보가 계쏙 표시됨
    folium.Marker([lat,lon], tooltip=name, popup=name, icon=folium.Icon(color=color)).add_to(map)
    
```

```python
data = pd.read_csv('sample/국토교통부_전국 버스정류장 위치정보_20241028.csv', encoding='euc-kr')
df = data[['정류장명', '위도', '경도', '도시명']]
df
```

```python
# 정류장 컬럼의 자료형은 object(string)
# 컬럼의 자료형이 string이면 str 속성으로 문자열함수를 사용 할 수 있다
df['정류장명'].str.contains('마을회관')
df[df['정류장명'].str.contains('마을회관')]
```

```python
df['도시명'] == '인천광역시'
result = df[df['도시명'] == '인천광역시']
df1 =result[result['정류장명'].str.contains('마을회관')]
```

서울 정류장에서 '상가'가 포함된 정류장만 선택

```python
df['도시명'] == '서울특별시'
seoul = df[df['도시명'] == '서울특별시']
seoul['정류장명'].str.contains('상가')
df2 = seoul[seoul['정류장명'].str.contains('상가')]
```

```python
df1
df2
bus = pd.concat([df1, df2])
bus
```

```python
map = folium.Map(location=[37.492642,126.857774], zoom_start=13, tooltip='정류장')


```

```python
folium.Marker([37.492642, 126.857774], tooltip='정류장').add_to(map)
```

```python
for i in bus.index:
    name = bus.loc[i, '정류장명']
    lat = bus.loc[i, '위도']
    lon = bus.loc[i, '경도']
    city = bus.loc[i, '도시명']
    if city=='서울특별시':
        color = 'blue'
    elif city=='인천광역시':
        color = 'red'

folium.Marker([lat, lon], tooltip=name, icon=folium.Icon(color=color)).add_to(map)
```
